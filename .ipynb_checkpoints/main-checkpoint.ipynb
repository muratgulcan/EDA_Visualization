```python
import pandas as pd
import numpy as np
```

```python
fileName = "https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/winequalityN.csv"
df = pd.read_csv(fileName)
df.head()
```

## Simple Descriptive Analysis

```python
df.describe()
```

```python
df.info()
```

```python
df['quality'].mean()
```

## Plotting using a Histogram

```python
df.hist(column='quality')
```

```python
# Let's get a bit more advanced learn to customize your plots

df.hist(column='quality', bins = 10, grid = True, figsize = (10,6), color = 'cornflowerblue')
```

These look nice but using the library Seaborn, we can quickly produce even nicer plots

```python
import seaborn as sns
```

```python
sns.displot(df.alcohol, bins = 25, kde = True)
```

```python
df.alcohol.value_counts().head()
```

## Styling and Axis Labels

```python
import matplotlib.pyplot as plt
```

```python
sns.displot(df.pH, kde = True)
plt.xlabel('pH Percentage')
plt.ylabel('Count')
plt.title('pH Content')
plt.show()
```

```python
# Use sns.set_style() to control styling

sns.set_style("dark")
sns.displot(df.pH, bins = 15, kde = True)
```

## Scatter Plots

```python
sns.lmplot(x='pH', y='fixed acidity', data=df)
```

```python
sns.lmplot(x='pH', y='density', data=df, fit_reg=False)
```

```python
# Scatterplot arguments

sns.lmplot(x='pH', y='chlorides', data = df, fit_reg = False, 
           hue= 'type' # color by type
          )
```

## Boxplots & Finding Outliers



```python
print(df.alcohol.quantile(([0.25,0.75])))
sns.boxplot(data=df.alcohol)
```

```python
sns.boxplot(data=df, palette='Set3')
```

```python
plt.figure(figsize = (12,8))
summary = df.drop(['free sulfur dioxide', 'total sulfur dioxide', 'quality', 'residual sugar', 
                  'fixed acidity', 'alcohol'], axis = 1)
sns.boxplot(data=summary, palette='Set3')
```

## Violin Plots

```python
sns.violinplot(data=df.pH)
```

## Bar Plot

```python
plt.figure(figsize=(12,8))
sns.countplot(x='quality', data=df, hue='type')
```

## Titanic Example

```python
# load the dataset
titanic = sns.load_dataset('titanic')

# draw a nested barplot to show survival for class and sex
g = sns.catplot(x='class', y='survived', hue='sex', data=titanic, height=5, kind='bar', palette='muted')

g.despine(left=True)
g.set_ylabels('survival probability')
```

## Density Plot

```python
sns.kdeplot(df.alcohol,df.density)
```

## Joint Distribution Plot

```python
sns.jointplot(x='alcohol', y = 'density', data=df)
```

## Factor Plots and Bee Swarm Plots
Factor plots allow us to create seperate plot by categorical classes

```python
# Factor plot
g = sns.factorplot(
    x='free sulfur dioxide',
    y='alcohol',
    data=df,
    hue='quality', #color by stage
    col='quality', #seperate by stage
    kind='point', #swarmplot
    height = 6
                  )
g.set_xticklabels(rotation=45)
```

## Creating Empirical cumulative distribution function (ECDF)

```python
x = np.sort(df.alcohol)
y = np.arange(1, len(x)+1) / len(x)
_ = plt.plot(x,y, marker='.', linestyle = 'none')
plt.margins(0.05)
_ = plt.xlabel('Percent of Alcohol in Wine')
_ = plt.ylabel('ECDF')
plt.grid(True)
plt.show()
```
